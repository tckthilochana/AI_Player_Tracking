```python
#instead of detecting all objects, detect only person
import os
from ultralytics import YOLO

# Load model
model = YOLO("yolov8n.pt")

video_folder = "/content/drive/MyDrive/Colab_Notebooks/AI_Assignment2/videos"

# Process every video in the folder
for video in os.listdir(video_folder):

    if video.endswith(".mp4"):

        video_path = os.path.join(video_folder, video)

        print(f"Processing {video}...")

        model.predict(
            source=video_path,
            classes=[0],          # person class only
            conf=0.4,
            save=True,
            project="/content/drive/MyDrive/Colab_Notebooks/AI_Assignment2",
            name="YOLOv8n_Output",
            exist_ok=True
        )

print("Finished!")
```
